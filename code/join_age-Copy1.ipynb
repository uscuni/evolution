```python
import geopandas as gpd
import pandas as pd
import lonboard
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
buildings = gpd.read_parquet("/data/uscuni-ulce/data_product/germany_buildings.parquet")
buildings
```

```python
files= glob(f"/data/uscuni-evolution/germany_ages/DE*.csv")
files
```

```python
ages = []
for f in files:
    table = pd.read_csv(f)
    ages.append(table)
ages_df = pd.concat(ages)
ages_df
```

```python
#ages_df['geometry'] = ages_df['position'].apply(loads)
ages_df['year'] = ages_df['construction_year'].apply(lambda x: int(x[10:14]))
ages_df
```

```python
geoms = gpd.GeoSeries.from_wkt(ages_df.position, crs=3035).to_crs(4326)
```

```python
gdf = gpd.GeoDataFrame(ages_df[['year', 'id']], geometry=geoms)
```

```python
joined = buildings.sjoin(gdf, predicate='contains')
```

```python
joined.to_parquet('/data/uscuni-evolution/germany_ages/buildings_age.parquet')
```

```python
morphotopes = gpd.read_parquet("/data/uscuni-ulce/data_product/germany_morphotopes.parquet")
morphotopes
```

```python
morphotopes.crs
```

```python
gdf.crs
```

```python
joined_morphotopes = gdf.sjoin(morphotopes[["geometry"]],predicate="within")
joined_morphotopes
```

```python
grouped = joined_morphotopes.groupby("morph")
df_size = grouped.size().reset_index()
df_size.columns = ['morph', 'num_buildings']
df_size
```

```python
morphotopes.to_parquet('/data/uscuni-evolution/germany_ages/morphotopes_age.parquet')
```

```python
grouped_stats = grouped.agg(
    num_buildings=('year', 'count'),  # <-- total number of buildings (rows)
    
    year_mean=('year', 'mean'),
    year_median=('year', 'median'),
    year_min=('year', 'min'),
    year_q10=('year', lambda x: x.quantile(0.1)),
    year_q90=('year', lambda x: x.quantile(0.9)),
    year_max=('year', 'max'),
    year_std=('year', 'std'),

).round(2).reset_index()
grouped_stats
```

```python
# Make sure 'morph' is a column in both DataFrames
#morphotopes = morphotopes.reset_index()

# Keep only morphs that exist in both DataFrames
morphotopes_joined = grouped_stats.merge(morphotopes, on='morph', how='inner')
morphotopes_joined
```
