```python
import geopandas as gpd
import pandas as pd
import lonboard
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
files_age = glob(f"/data/uscuni-evolution/germany_ages/DE*.csv")
files_age
```

```python
ages = []
for f in files_age:
    table = pd.read_csv(f)
    ages.append(table)
ages_df = pd.concat(ages)
ages_df
```

```python
#ages_df['geometry'] = ages_df['position'].apply(loads)
ages_df['year'] = ages_df['construction_year'].apply(lambda x: int(x[10:14]))
ages_df
```

```python
geoms = gpd.GeoSeries.from_wkt(ages_df.position, crs=3035)
```

```python
gdf = gpd.GeoDataFrame(ages_df[['year', 'id']], geometry=geoms)
gdf
```

```python
files_morph = glob(f"/data/uscuni-ulce/data_product/de*_morphotopes.parquet")
morphotopes_list = []
for f in files_morph:
    data = gpd.read_parquet(f)
    morphotopes_list.append(data)
morphotopes = pd.concat(morphotopes_list).set_index("morphotope_id")
morphotopes
```

```python
morphotopes_no_noise = morphotopes[~morphotopes.initially_noise]
```

```python
morphotopes_no_noise.index.is_unique
```

```python
joined_morphotopes = gdf.sjoin(morphotopes_no_noise[["geometry"]], predicate="within")
joined_morphotopes
```

```python
bins = (0,1919,1948,1978,1986,1990,1995,2000,2004,2008,2030)
def mode_bin(x):
    binned = pd.cut(x, bins=bins, right=True)
    m = binned.mode()
    return m.iloc[0] if not m.empty else None
```

```python
grouped = joined_morphotopes.groupby("morphotope_id")
```

```python
grouped_stats = grouped.agg(
    num_buildings=('year', 'count'),  # <-- total number of buildings (rows)

    year_mean=('year', 'mean'),
    year_median=('year', 'median'),
    year_min=('year', 'min'),
    year_q10=('year', lambda x: x.quantile(0.1)),
    year_q90=('year', lambda x: x.quantile(0.9)),
    year_max=('year', 'max'),
    year_std=('year', 'std'),
    year_mode=('year', mode_bin)
).round(2)
grouped_stats
```

```python
grouped_stats.index.is_unique
```

```python
# Keep only morphs that exist in both DataFrames
morphotopes_age = morphotopes_no_noise.join(grouped_stats, how='inner')
morphotopes_age
```

```python
morphotopes_age.to_parquet('/data/uscuni-evolution/germany_ages/morphotopes_age.parquet')
```
