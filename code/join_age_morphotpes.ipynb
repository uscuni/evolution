```python
import geopandas as gpd
import pandas as pd
import lonboard
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
files= glob(f"/data/uscuni-ulce/germany_ages/DE*.csv")
files
```

```python
ages = []
for f in files:
    table = pd.read_csv(f)
    ages.append(table)
ages_df = pd.concat(ages)
ages_df
```

```python
#ages_df['geometry'] = ages_df['position'].apply(loads)
ages_df['year'] = ages_df['construction_year'].apply(lambda x: int(x[10:14]))
ages_df
```

```python
geoms = gpd.GeoSeries.from_wkt(ages_df.position, crs=3035).to_crs(4326)
```

```python
gdf = gpd.GeoDataFrame(ages_df[['year', 'id']], geometry=geoms)
```

```python
gdf
```

```python
morphotopes = gpd.read_parquet("/data/uscuni-ulce/data_product/germany_morphotopes.parquet")
morphotopes
```

```python
morphotopes.index.is_unique
```

```python
joined_morphotopes = gdf.sjoin(morphotopes[["geometry"]],predicate="within")
joined_morphotopes
```

```python
grouped = joined_morphotopes.groupby("morph")
```

```python
grouped_stats = grouped.agg(
    num_buildings=('year', 'count'),  # <-- total number of buildings (rows)
    
    year_mean=('year', 'mean'),
    year_median=('year', 'median'),
    year_min=('year', 'min'),
    year_q10=('year', lambda x: x.quantile(0.1)),
    year_q90=('year', lambda x: x.quantile(0.9)),
    year_max=('year', 'max'),
    year_std=('year', 'std'),

).round(2).reset_index()
grouped_stats
```

```python
# Make sure 'morph' is a column in both DataFrames
morphotopes = morphotopes.reset_index()

# Keep only morphs that exist in both DataFrames
morphotopes_age = grouped_stats.merge(morphotopes, on='morph', how='inner')
morphotopes_age
```

```python
morphotopes_age.to_parquet('/data/uscuni-ulce/germany_ages/morphotopes_age.parquet')
```
