```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.cm as cm
from glob import glob
```

Buildings with ages and height

```python
!ls /data/uscuni-evolution/
```

```python
files_age = glob(f"/data/uscuni-evolution/germany_ages/DE*.csv")
files_age
```

```python
ages = []
for f in files_age:
    table = pd.read_csv(f)
    ages.append(table)
buildings = pd.concat(ages)
buildings
```

```python
buildings['year'] = buildings['construction_year'].apply(lambda x: int(x[10:14]))
buildings
```

```python
geoms = gpd.GeoSeries.from_wkt(buildings.position, crs=3035)
gdf = gpd.GeoDataFrame(buildings[['year']], geometry=geoms)
gdf
```

Morphotopes.

```python
files_morph = glob(f"/data/uscuni-ulce/data_product/de*_morphotopes.parquet")
morphotopes_list = []
for f in files_morph:
    data = gpd.read_parquet(f)
    morphotopes_list.append(data)
morphotopes = pd.concat(morphotopes_list).set_index("morphotope_id")
morphotopes
```

Join buidings to morphtopes

```python
morphotopes=morphotopes[morphotopes["initially_noise"]==False]
```

```python
joined_morphotopes = gdf.sjoin(morphotopes[["geometry"]], predicate="within")
joined_morphotopes
```

Group buildings per morp

```python
bins = (0,1919,1948,1978,1986,1990,1995,2000,2004,2008,2030)
def mode_bin(x):
    binned = pd.cut(x, bins=bins, right=True)
    m = binned.mode()
    return m.iloc[0] if not m.empty else None
```

```python
grouped = joined_morphotopes.groupby("morphotope_id")
```

```python
grouped_stats = grouped.agg(
    num_buildings=('year', 'count'),  # <-- total number of buildings (rows)

    year_mean=('year', 'mean'),
    year_median=('year', 'median'),
    year_min=('year', 'min'),
    year_q10=('year', lambda x: x.quantile(0.1)),
    year_q90=('year', lambda x: x.quantile(0.9)),
    year_max=('year', 'max'),
    year_std=('year', 'std'),
    year_mode=('year', mode_bin)
).round(2)
grouped_stats
```

```python
morphotopes_age = morphotopes.join(grouped_stats, how='inner')
morphotopes_age
```

```python
morphotopes_age = morphotopes_age[
    (morphotopes_age['num_buildings'] != 0) &
    (morphotopes_age['initially_noise'] == False)
]
```

```python
morphometrics = pd.read_parquet("/data/uscuni-ulce/data_product/morphotope_data.parquet")
morphometrics
```

```python
joined = morphotopes_age.join(morphometrics,how='left')
joined
```

```python
joined.to_parquet("/data/uscuni-evolution/all_joined_de.parquet")
```
